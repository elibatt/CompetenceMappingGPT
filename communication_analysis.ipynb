```python
%config InlineBackend.figure_formats = ['retina']

import pandas as pd
from sklearn.metrics import mean_squared_error
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

com_results_first = pd.read_csv('data/results_communication_competences_v1.0.csv')
com_results_second = pd.read_csv('data/results_communication_competences_v2.0.csv')
com_results_third = pd.read_csv('data/results_communication_competences_v2.1.csv')
com_results_final = pd.read_csv('data/results_communication_competences_v1.1.csv')

dict = {"v1.0": com_results_first, "v1.1": com_results_final, "v2.0": com_results_second, "v2.1": com_results_third}

values = []

for key in dict.keys():
    data = dict[key]
    actual = data["TARGET_OVERALL"]
    predicted = data["OVERALL"]
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mse)
    print(f"{key}: {rmse}")
    values.append(rmse)

sns.set(rc={"figure.dpi":100, 'savefig.dpi':100})
sns.set_context('notebook')
sns.set_style("ticks")


plt.figure(figsize=(10, 6))
sns.barplot(x=list(dict.keys()), y=values, hue=list(dict.keys()))
plt.title("Overall RMSE")
plt.ylabel("RMSE Values")
plt.yticks(np.arange(0.0, max(values) + 0.2, 0.2))
sns.despine(left=False, bottom=False)
```

```python
values = []
keys = list(dict.keys())
num_lists = len(keys)
bar_width = 0.2

for key in keys:
    data = dict[key]
    actual = data["TARGET_OVERALL"]
    predicted = data["OVERALL"]
    values.append(abs(actual-predicted+0.025))

positions = np.arange(-bar_width, len(values[0])-1+bar_width, 1.0)

fig, ax = plt.subplots(figsize=(10, 6))

for i in range(num_lists):
    ax.bar(-0.1+positions + i * bar_width, values[i], bar_width, label=keys[i])

ax.legend()
plt.xticks(np.arange(0.0, len(values[0])-0.5, 1.0))
plt.yticks(np.arange(0.0, 3.01, 0.25))
plt.title("Error per transcription")
plt.ylabel("RMSE Values")
plt.xlabel("Transcriptions")
sns.despine(left=False, bottom=False)
```
